# Data description
● city: city this user signed up in

● phone: primary device for this user

● signup_date: date of account registration; in the form ‘YYYYMMDD’

● last_trip_date: the last time this user completed a trip; in the form ‘YYYYMMDD’

● avg_dist: the average distance in miles per trip taken in the first 30 days after signup

● avg_rating_by_driver: the rider’s average rating over all of their trips

● avg_rating_of_driver: the rider’s average rating of their drivers over all of their trips

● surge_pct: the percent of trips taken with surge multiplier > 1

● avg_surge: The average surge multiplier over all of this user’s trips

● trips_in_first_30_days: the number of trips this user took in the first 30 days after signing up

● ultimate_black_user: TRUE if the user took an Ultimate Black in their first 30 days; FALSE
otherwise

● weekday_pct: the percent of the user’s trips occurring during a weekday

# Missing Value Imputation

The missingness matrix doesn't reveal any apparent pattern in the missing data, although there are a significant amount of missing observations for the avg_rating_of_driver feature (roughly 17%).

```python
msno.matrix(df)
```

The avg_rating_of_driver and avg_rating_by_driver features were imputed with the rounded mean value of the column. The phone categorical feature was imputed with the mode value.

```python
df[['avg_rating_of_driver']] = df[['avg_rating_of_driver']].fillna(df[['avg_rating_of_driver']].mean().round())
df[['avg_rating_by_driver']] = df[['avg_rating_by_driver']].fillna(df[['avg_rating_by_driver']].mean().round())
df[['phone']] = df[['phone']].fillna(df[['phone']].mode().iloc[0])
df
```

# Feature Engineering

It was necessary to convert some of the date features into integer values representing the time elapsed since sign-up and last_trip_date.

```python
df['signup_date'] = pd.to_datetime(df['signup_date'], infer_datetime_format=True)
df['last_trip_date'] = pd.to_datetime(df['last_trip_date'], infer_datetime_format=True)
df
```

I also had to create the target categorical feature - whether or not the user can be considered a retained user. This feature was created based on the definition from the scope: "we consider a user retained if they were “active” (i.e. took a trip) in the preceding 30 days."

```python
df['retained'] = 0
df.loc[df.index.isin(retained.index), 'retained'] = 1
for column_name in ['city', 'phone', 'ultimate_black_user', 'retained']:
    df[column_name] = df[column_name].astype('category')
df
```

```python
retained_max = df['last_trip_date'].max()
retained_min = retained_max - timedelta(days=30)
retained = df.loc[(df['last_trip_date'] <= retained_max) & (df['last_trip_date'] >= retained_min)]
```

```python
df['signup_date_days'] = datetime.now() - df.signup_date
df['signup_date_days'] = df['signup_date_days'].dt.days
df['last_trip_date_days'] = datetime.now() - df.last_trip_date  
df['last_trip_date_days'] = df['last_trip_date_days'].dt.days
df = df.drop(columns=df.select_dtypes(['datetime']).columns)
df
```

# Preprocessing

## One Hot Encoding

```python
df = pd.get_dummies(df, drop_first=True)
df
```

## Train-Test Split

```python
X = df.drop(columns = ['retained_1', 'last_trip_date_days'])
y = df['retained_1']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

# Model Development

## Gradient Boosting Classifier

```python
param_grid = dict(
    learning_rate=[0.001, 0.05, 0.1, 0.2, 0.5],
    max_depth=list(range(4,10)),
    min_samples_leaf=list(range(1,21)),
    min_samples_split=list(range(5,51)),
)
gb = GradientBoostingClassifier(n_estimators=50, max_features=2, 
                                random_state = 42)
gb_cv = BayesSearchCV(gb, param_grid, scoring='f1', cv=5, n_iter = 32)
gb_cv.fit(X_train, y_train)
```

```python
gb_cv.best_params_
```

```python
y_pred = gb_cv.predict(X_train)
```

```python
print(classification_report(y_train, y_pred))
```

# Model Evaluation

```python
print(classification_report(y_test, y_pred))
```
