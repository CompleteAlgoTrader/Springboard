#### 1.1. Importing Libraries

```python
# Let's import the pandas, numpy libraries as pd, and np respectively. 
import pandas as pd
import numpy as np

# Load the pyplot collection of functions from matplotlib, as plt 
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices= "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
properties = properties.T.sort_index()
properties.columns = properties.loc['Unnamed: 0']
properties = properties.reset_index()
print(properties.head())
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
properties.rename(columns = {'index':'Borough', pd.NaT:'Post Code'}, inplace = True)
properties.rename_axis('Index', axis = 1, inplace = True)
print(properties.head())
```

```python
unnamed_list = ['Unnamed: 0', 'Unnamed: 34', 'Unnamed: 37', 'Unnamed: 47']
unnamed_rows = properties.loc[properties['Borough'].isin(unnamed_list)]

not_a_borough_list = ['City of London', 'EAST MIDLANDS', 'EAST OF ENGLAND', 'England', 'Inner London', 'LONDON', 'NORTH EAST', 'NORTH WEST', 'Outer London', 'SOUTH EAST', 'SOUTH WEST', 'WEST MIDLANDS', 'YORKS & THE HUMBER']
not_a_borough_rows = properties.loc[properties['Borough'].isin(not_a_borough_list)]
properties = properties.drop(unnamed_rows.index)
properties = properties.drop(not_a_borough_rows.index)
print(properties)
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
properties = properties.melt(id_vars = ['Borough', 'Post Code'], value_name = 'Average Price', var_name = 'Date').reset_index(drop = True)
properties['Year'] = [date.year for date in properties['Date']]
properties['Month'] = [date.month for date in properties['Date']]
properties['Day'] = [date.day for date in properties['Date']]
properties.sort_values(by = ['Borough', 'Date'])
print(properties)
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
print(type(properties['Average Price'][0]))
print(type(properties['Date'][0]))
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
print(f"There are {properties[properties['Borough'].isnull() == True]['Borough'].count()} null rows in my DataFrame.")
print(f"\nThere are {properties['Borough'].nunique()} boroughs in the DataFrame.")
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
last_twenty_years = properties.query("Year >= 2001")['Date']
last_twenty_years_prices = properties.query("Year >= 2001")['Average Price']
_ = enumerate(last_twenty_years, start = last_twenty_years.index[0])
x_tick_label = {}
for i,j in _:
    if j.year not in list(x_tick_label.values()):
        x_tick_label[i] = j.year

fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

for borough in properties['Borough'].unique():
    plt.plot(properties.query(f"Year >= 2001 and Borough == '{borough}'")['Average Price'])
plt.title('London Borough Housing Prices 2001-2021')
plt.xlabel('Year')
plt.ylabel('Avg. Property Value')
plt.xticks(list(x_tick_label.keys()), list(x_tick_label.values()), rotation = 45)
plt.yticks([200000, 400000, 600000, 1000000, 1400000], ['200,000', '400,000', '600,000', '1,000,000', '1,400,000'], rotation = 45)
plt.tick_params(labelsize = 10)
plt.legend(list(properties['Borough'].unique()), bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2021 to the price in 2001.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 2001 and 2021.
4. Calculate the ratio of the Average Price for 2001 divided by the Average Price for 2021.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 2001 and 2021.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
create_price_ratio = lambda price1, price2: price1 / price2
twenty_year_prices = pd.DataFrame()
twenty_year_prices['Borough'] = properties.query("Year == 2001 and Month == 1")['Borough'].reset_index(drop = True)
twenty_year_prices['Historical'] = properties.query("Year == 2001 and Month == 1")['Average Price'].reset_index(drop = True)
twenty_year_prices['Current'] = properties.query("Year == 2021 and Month == 1")['Average Price'].reset_index(drop = True)
twenty_year_prices['Ratio'] = create_price_ratio(twenty_year_prices['Historical'], twenty_year_prices['Current'])
twenty_year_prices.set_index('Borough', inplace = True)

print(twenty_year_prices.query(f"Current == {twenty_year_prices['Current'].max()}"))
```

```python
print(twenty_year_prices.query(f"Ratio == {twenty_year_prices['Ratio'].min()}"))

CAGR = lambda price1, price2, years: ((float(price2) / float(price1)) ** (1/years) - 1) * 100
_ = []
for borough in twenty_year_prices.index:
    _.append(CAGR(properties.query(f"Year == 2001 and Month == 1 and Borough == '{borough}'")['Average Price'], properties.query(f"Year == 2021 and Month == 1 and Borough == '{borough}'")['Average Price'], 20))
twenty_year_prices['CAGR'] = _
print(twenty_year_prices.query(f"CAGR == {twenty_year_prices['CAGR'].max()}"))
```

```python
fig = plt.figure(figsize=(15,8))

ax = fig.add_subplot(111)
plt.title('Waltham Forest 20 Year Avg. Housing Prices',fontsize=19,ha='center')
plt.plot('Average Price', data=properties.query("Borough == 'Waltham Forest' and Year >= 2001"))
plt.grid(axis='y')
plt.xticks(list(x_tick_label.keys()), list(x_tick_label.values()), rotation = 45)
plt.xlabel('Year')
plt.ylabel('Avg. Property Value')
plt.show()
```
